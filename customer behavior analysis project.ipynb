```python
#Here we are importing pandas libraries that is pandas as pd. and we declaring variable named df.
#The pd.read_csv function is used to upload the csv file and used head function to view the few rows of data.
import pandas as pd
df=pd.read_csv('customer_shopping_behavior.csv')
df.head()
```

```python
'''Q.what is df.info()?

Ans. df.info() is one of the most important functions in pandas

It is used to:

Get a complete summary of your DataFrame'''
df.info()
```

```python
'''df.describe() it is used for find complete statistical values only for

numerical columns. and include='all' which makes to find the complete

statical data of all numerical and categorical columns.'''

df.describe(include='all')
```

```python
'''is null() is a function used to find null values in dataframe and .sum()

functions is used to to count missing values'''

df.isnull().sum()
```

```python
'''taking the column name of Reviw Rating and grouped by all category of Review

Rating and using the function of transform function inside that lambda

function is used by giving value of x .and x means all values of column

taking the median of all value and filling the null values in the column.'''

df['Review Rating']=df.groupby('Category')['Review Rating'].transform(lambda x:x.fillna(x.median))
```

```python
'''again checking for the null values it showing as no null values'''

df.isnull().sum()
```

```python
'''making the all columns into lower case using the lower function by str

method. and making all column replacing using replace function the space

into underscore and printing the column names.'''

df.columns =df.columns.str.lower()
df.columns =df.columns.str.replace(' ','_')
df.columns
```

```python
'''so here is the purchase_amount_(usd) this column name contain the (usd) in

the in braket so it is replacing as purchase_amount because it looks like

complex name.'''

df=df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
df.columns
```

```python
'''for the buisness analysis we are creating the another column name called

age_group so we are creating four labels as Young_adult ,Adult,Middle_aged,

Senior. age_group the column name is declared newly but due to the pandas it

autmiticalliy creates the column name. here the pd.qcut function makes the

four parts of the quartiles and we are printing the age and age_group column names.'''

#creating age into age_group
labels=['Young_adult','Adult','Middle_aged','Senior']
df['age_group']=pd.qcut(df['age'],q=4,labels=labels)
df[['age','age_group']].head(10)
```

```python
'''here checking the unique category of the column names by unique function to map the numerical values instead of the catogical data.'''

#create column of purchase_frequency_days

frequency_mapping ={
    'Fortnightly':14,
    'Weekly':7,
    'Annually':365,
    'Quarterly':90,
    'Monthly':30,
    'Bi-Weekly':14,
    'Every 3 Months':90
}
df['purchase_frequency_days']=df['frequency_of_purchases'].map(frequency_mapping)
```

```python
'''checking the values are proper by using the head function.'''

df[['purchase_frequency_days','frequency_of_purchases']].head(10)
```

```python
'''here the discount _applied and the promo_code_used are linking in buiseness

because when some one promote to buy product means customer will have chance of

getting discount so we are using to view the data.'''

df[['discount_applied','promo_code_used']].head(10)
```

```python
'''we are checking the values both are true.'''
df['discount_applied'] == df['promo_code_used']
```

```python
'''here we are checking all are same by comparing two column names and applying

the all() fuction to check with all values.'''

bool((df['discount_applied']==df['promo_code_used']).all())
```

```python
'''so here two column names are same so we are dropping the promo_code_used column

name where axis is 1 means the one column name will be deleted.'''

df=df.drop('promo_code_used',axis=1)
```

```python
'''and printing the coloumn names which are updated.'''

df.columns
```
